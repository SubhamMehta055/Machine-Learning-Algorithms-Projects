## PCA [Dimensionality Reduction]

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Plotting the results
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=plt.get_cmap('viridis', 3))
plt.colorbar(scatter, ticks=[0, 1, 2], label='Class')
#no. of components can be <= no. of features 
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
'''
Applications of PCA :

Visualization: Simplify high-dimensional data to 2D or 3D for visualization.
Preprocessing: Reduce dimensionality before applying machine learning algorithms to improve performance and reduce overfitting.
Feature Extraction: Identify important features and eliminate redundant or less informative ones.
```
