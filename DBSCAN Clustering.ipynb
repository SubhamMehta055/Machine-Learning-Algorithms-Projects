## DBSCAN CLUSTERING

make_moons generates a two-dimensional dataset with two interleaving half circles. Itâ€™s useful for testing clustering algorithms on data with non-linear shapes.
We generate 300 samples with noise=0.1 to add some randomness to the data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN  #useful for non-linear structure
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler

# Generate the make_moons dataset
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# Optionally, scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)  #first fitting and then directly returns the cluster labels or predictions for the input data

# Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_dbscan, cmap='rainbow', s=50)
plt.title("DBSCAN Clustering on make_moons Dataset")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
'''
Key Parameters of DBSCAN:
eps: Maximum distance between two samples for them to be considered as neighbors.
min_samples: The minimum number of samples in a neighborhood for a point to be considered a core point.
Labels:
Cluster labels: 0, 1, 2, etc.
Outliers (Noise): Labeled as -1.
'''
```

This implementation uses DBSCAN to handle the non-linear structure of the make_moons dataset, and you'll see how DBSCAN can form clusters even when the shapes are not circular.

### DBSCAN is very useful when your data is in non-linear relationship and helps in clustering in effective way
